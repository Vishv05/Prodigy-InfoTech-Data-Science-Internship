```python
# ============================================================
# Task 01: Distribution Visualization
# Internship: Prodigy InfoTech â€“ Data Science
# Intern Name: Vishv Bhavsar
# Track Code: Data Science
# ============================================================

# ---------------------------
# 1. Import Required Libraries
# ---------------------------
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

# ---------------------------
# 2. Load Dataset Files
# ---------------------------
# Main Population Data (Skip first 4 metadata rows)
population_df = pd.read_csv(
    "API_SP.POP.TOTL_DS2_en_csv_v2_38144.csv",
    skiprows=4
)

# Metadata Files (Reference Only)
country_metadata = pd.read_csv(
    "Metadata_Country_API_SP.POP.TOTL_DS2_en_csv_v2_38144.csv"
)

indicator_metadata = pd.read_csv(
    "Metadata_Indicator_API_SP.POP.TOTL_DS2_en_csv_v2_38144.csv"
)

# ---------------------------
# 3. Basic Dataset Overview
# ---------------------------
print("Population Dataset Shape:", population_df.shape)
print("\nColumns Available:")
print(population_df.columns[:10])

print("\nFirst 5 Rows:")
display(population_df.head())

# ---------------------------
# 4. Select Year for Analysis
# ---------------------------
year = "2020"

# Extract only relevant columns
yearly_population = population_df[["Country Name", year]].dropna()

print(f"\nTotal Countries with Population Data in {year}:",
      yearly_population.shape[0])

# ---------------------------
# 5. Histogram â€“ Continuous Variable
# ---------------------------
plt.figure(figsize=(10, 6))
sns.histplot(yearly_population[year], bins=30, kde=True)

plt.title(f"Distribution of Population Across Countries ({year})")
plt.xlabel("Population")
plt.ylabel("Number of Countries")

plt.show()

# ---------------------------
# 6. Bar Chart â€“ Top 10 Most Populated Countries
# ---------------------------
top_10 = yearly_population.sort_values(by=year, ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=year, y="Country Name", data=top_10)

plt.title(f"Top 10 Most Populated Countries ({year})")
plt.xlabel("Population")
plt.ylabel("Country")

plt.show()

# ---------------------------
# 7. Observations
# ---------------------------
print("""
Observations:
1. The population distribution is highly right-skewed.
2. Most countries have relatively smaller populations.
3. A few countries like China and India dominate global population.
4. The bar chart clearly shows population concentration among top countries.
""")

# ---------------------------
# 8. Conclusion
# ---------------------------
print("""
Conclusion:
The histogram and bar chart effectively visualize the distribution
of population across countries. This fulfills Task-01 requirement
of analyzing and visualizing a continuous variable distribution.
""")
```
