```python
# ==========================================================
# Task-02: Data Cleaning & Exploratory Data Analysis (EDA)
# Dataset: Titanic (train.csv)
# ==========================================================

# -----------------------------
# 1. Import Required Libraries
# -----------------------------
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

print("Libraries Imported Successfully\n")

# -----------------------------
# 2. Load Dataset
# -----------------------------
df = pd.read_csv("train.csv")

print("Dataset Loaded Successfully\n")

# -----------------------------
# 3. Basic Dataset Information
# -----------------------------
print("First 5 Rows:\n")
print(df.head())

print("\nDataset Shape:", df.shape)

print("\nDataset Info:\n")
print(df.info())

print("\nStatistical Summary:\n")
print(df.describe())

# -----------------------------
# 4. Check Missing Values
# -----------------------------
print("\nMissing Values:\n")
print(df.isnull().sum())

# -----------------------------
# 5. Data Cleaning
# -----------------------------

# Fill missing Age values with median
df['Age'].fillna(df['Age'].median(), inplace=True)

# Fill missing Embarked values with mode
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)

# Drop Cabin column (too many missing values)
df.drop('Cabin', axis=1, inplace=True)

print("\nMissing Values After Cleaning:\n")
print(df.isnull().sum())

# -----------------------------
# 6. Survival Count
# -----------------------------
plt.figure()
sns.countplot(x='Survived', data=df)
plt.title("Survival Count")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.show()

# -----------------------------
# 7. Survival by Gender
# -----------------------------
plt.figure()
sns.countplot(x='Survived', hue='Sex', data=df)
plt.title("Survival by Gender")
plt.show()

# -----------------------------
# 8. Survival by Passenger Class
# -----------------------------
plt.figure()
sns.countplot(x='Survived', hue='Pclass', data=df)
plt.title("Survival by Passenger Class")
plt.show()

# -----------------------------
# 9. Age Distribution
# -----------------------------
plt.figure()
sns.histplot(df['Age'], bins=30, kde=True)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()

# -----------------------------
# 10. Fare Distribution
# -----------------------------
plt.figure()
sns.histplot(df['Fare'], bins=30, kde=True)
plt.title("Fare Distribution")
plt.xlabel("Fare")
plt.ylabel("Frequency")
plt.show()

# -----------------------------
# 11. Survival by Age
# -----------------------------
plt.figure()
sns.boxplot(x='Survived', y='Age', data=df)
plt.title("Survival by Age")
plt.show()

# -----------------------------
# 12. Correlation Heatmap
# -----------------------------
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.title("Correlation Heatmap")
plt.show()

# -----------------------------
# 13. Survival Rate Calculations
# -----------------------------
print("\nOverall Survival Rate:", df['Survived'].mean())

print("\nSurvival Rate by Gender:")
print(df.groupby('Sex')['Survived'].mean())

print("\nSurvival Rate by Passenger Class:")
print(df.groupby('Pclass')['Survived'].mean())

# -----------------------------
# 14. Final Observations
# -----------------------------
print("\nKey Observations:")
print("1. Females had significantly higher survival rate than males.")
print("2. 1st class passengers had highest survival probability.")
print("3. Higher fare passengers had better survival chances.")
print("4. Socio-economic status played a major role in survival.")
print("\nTask-02 Completed Successfully!")
```
